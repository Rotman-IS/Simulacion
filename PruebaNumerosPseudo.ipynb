## Método de cuadrados Medios

4. Generar un histograma con el resultado obtenidos por cada método.

```python
import matplotlib.pyplot as plt
diccionario = {i:contadorInteraciones.count(i) for i in contadorInteraciones}

X = np.arange(len(diccionario))

plt.figure(figsize=(14,6))
plt.bar(X, diccionario.values(), align='center', width=0.5, color='g')
plt.xticks(X, diccionario.keys())
ymax = max(diccionario.values()) + 1
plt.ylim(0, ymax)
plt.show()
```

## Método de Congruencias Lineales.


```python
import numpy as np
import time


rango = int((input("Ingrese la cantidad de numeros pseudoaleatorios:  ")))

a = 5
b = 3
m = 16

cont = 0
lista = []
contInterac = []

for n in range(rango):
    semilla = int(time.time())
    
    print("semilla",semilla,"iteracion",n)
    cont = 0
    while(True):
        
        xn = (a * semilla + b) % m
        semilla = xn 
        div = semilla / m
        
        #print(cont,'\t\t',semilla,'\t',"{0:.4f}".format(div),(semilla in lista) is True)
        
        if((div in lista) is True):
            
            contInterac.append(cont)
            
            break
       
        cont = cont+1
        lista.append(div)
        
    lista = []
    
```

### Grafica 

```python
import matplotlib.pyplot as plt

dic = {i:contInterac.count(i) for i in contInterac}



X = np.arange(len(dic))

plt.figure(figsize=(14,6))
plt.bar(X, dic.values(), align='center', width=0.5, color='g')
plt.xticks(X, dic.keys())
ymax = max(dic.values()) + 1
plt.ylim(0, ymax)
plt.show()
```
