## Aplicación Word.  
Acontinuacion se mostrara los comandos basicos para realizar procesos de escritura y lectura de un documento Word, tambiense mostrara comandos que nos permitira abrir archivos Word para poder modificarlos y asu vez exportarlo con otros formatos como (html,PDF.etc).

1. **Comando Word() -->** Este comando nos permite iniciar un documento de Word.
2. **Comando append_text(texto)** --> Este comando nos permite agregar un texto al documento de Word.

```python
from automagica import *

word = Word()
word.append_text('Hola Mundo! realizado desde la herramienta Automagica')
```

### Resultado de los comando anteriormente usados.
  

3. **Comando replace_text(palabra, nuevaPalabra)** --> El siguiente comando nos permite remplazar una palabra dentro de un documento de Word.
4. **read_all_text(return_as_list=False)** --> Este comando nos permite Leer todo el texto que contiene nuestros documeto Word.

```python
print("Lectura del texto original","==>", word.read_all_text())

word.replace_text('Automagica', 'Automagica-RPA')

print("Lectura del texto Modificado","==>",word.read_all_text())
```

Los siguientes comandos nos permitirán agregar una cabecera y un pie de pagina a nuestro documento de Word. 
5. **Comando set_headers(text)** --> El siguiente comando nos permite agregar una cabecera con el texo quese le pase como parametro.
6. **Comando set_footers(text)** --> El siguiente comando nos permite agregar una pie de pagina con el texo quese le pase como parametro.

```python
word.set_headers('¡Esta es una cabecera agregada con Automatica!')
word.set_footers('¡Esta es una pie de pagina agregada con Automatica!')
```

### Resultado obtenido con los comandos anteriores.
  

A continuacion, se muestra los comandos necesarios para exportar nuestros archivos word a html y pdf.
7. **Comando export_to_html()** --> Este comando nos permite realizar la convercion de word a html.
8. **Comando export_to_pdf()** --> Este comando nos ptemite realizar la convercion de word a pdf.

## Aplicación Excel.  
Acontinuacion se mostrara los comandos basicos para realizar procesos de escritura y lectura de un documento Excel, tambien se mostrara comandos que nos permitira abrir archivos Excel para poder modificarlos y asu vez exportarlo con otros formatos como (PDF).

A continuacion, se mostrara los cmandos que nospermitira crear un nuevo archivo de excel y agregar informacin en sus celdas.  
1. **Comando Excel()** --> Nos permite crear un documento nuevo de Excel.
2. **Comando write_cell()** --> Este comando nos permite agregar informacion a nuestro documento Excel.

```python
excel = Excel()
excel.write_cell(1,1,'Colum A')
excel.write_cell(2,1,'Colum B')
excel.write_cell(3,1,'Colum C')
```

### Resultado de los comandos anteriores.  
  
Los siguientes comandos nos permite activar una celda dependiendo de la informacion que se posea.  

3. **Comando activate_first_empty_cell_down()**
4. **Comando activate_first_empty_cell_up()**
5. **Comando activate_first_empty_cell_left()**
6. **Comando activate_first_empty_cell_right()**

```python
excel.activate_first_empty_cell_down()# El uso de este comando possiciona o activa la celda A2.
```

  
7. **activate_range(range_)** --> este comando nos permite seleccionar un rango de celdas especificas.

```python
excel.activate_range('A1:C5')
```

### Resultado del comando anterior.  
  
8. **Comando add_worksheet()** --> Agrega una nueva hoja de trabajo.
9. **Comando activate_worksheet()** --> Activa una hoja de trabajo previamente creada.

```python
excel.add_worksheet('Ejemplo1')
excel.activate_worksheet('Ejemplo1')    
```

### Resultados de la creacion y activacion de una hoja de trabajo.  
  
Los comando que se mostraran a continuacion nos permite exportar nuestro documento excel a pdf.
10. **export_to_pdf()**

```python
excel.export_to_pdf('C:/Users/Zurita/Desktop/ROT/10CICLO/Simulacion/GIT-JupyterNotbook/Tarea/AutomagicaCap/SalidaExcel.pdf')
```

### Resultado de exportar a pdf.  
  
Continuando se mostrara los comandos necesario para leer nuestro documento excel.
11. **read_cell(column, row)**
12. **read_range(range_)**
13. **excel.read_worksheet()**

```python
print ("Lectura de una celda ",excel.read_cell(1,1))
print ("Lectura de un rango de celdas ", excel.read_range('A1:C5'))
print ("Lectura de la hoja de trabajo con nombre hoja1", excel.read_worksheet('Hoja1'))
```
